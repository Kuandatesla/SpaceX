```python
# TASK 3: Calculate the distances between a launch site to its proximities
import numpy as np

# Define a helper function using the Haversine formula
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth's radius in kilometers
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# Example: calculate distance from each launch to a reference site (e.g., KSC LC-39A)
ref_site = spacex_df[spacex_df['Launch Site'] == 'KSC LC-39A'].iloc[0]
ref_lat, ref_lon = ref_site['Lat'], ref_site['Long']

# Add a new column with distance to the reference site
spacex_df['distance_km_to_KSC'] = spacex_df.apply(
    lambda row: haversine(ref_lat, ref_lon, row['Lat'], row['Long']), axis=1
)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
